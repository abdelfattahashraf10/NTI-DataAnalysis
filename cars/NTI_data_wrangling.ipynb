## 4-iv Inconsistant Data Type

```python
df.dtypes
```

### iii-a Define
- columns Data Type issue
    - ['price',  'stroke', 'bore'] object insted of float
    - ['hoursepower', 'peak-rpm'] object insted of int64
    - ['num-of-cylinders', 'num-of-doors'] numbers in letters

### iii-b Code

#### convert to float

```python
#columns need to convert to float
columns_to_float = ['price', 'stroke', 'bore']

# Convert the columns to float, coercing errors to NaN
df[columns_to_float] = df[columns_to_float].apply(pd.to_numeric, errors='coerce')
```

#### convert to int

```python
#columns need to convert to int
columns_to_int = ['horsepower', 'peak-rpm']

# Convert the columns to int, coercing errors to NaN
df[columns_to_int] = df[columns_to_int].apply(pd.to_numeric, errors='coerce')
```

#### convert letters to numbers

```python
# know what are the values in column
print(df['num-of-cylinders'].unique())
```

```python
# Assign numerical values in ['num-of-cylinders'] column
cylinder_dic = {
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'eight': 8,
    'twelve': 12
}

# Replace the string values in the column with their corresponding integer values
df['num-of-cylinders'] = df['num-of-cylinders'].map(cylinder_dic)
df['num-of-cylinders']
```

# 7- Visualization
- i- Distribution of Car Prices
- ii- Horsepower vs. Price

## i- Distribution of Car Prices

```python
sns.histplot(df['price'], bins=20, kde=True, color='blue')
plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)

plt.show()
```

## ii- Horsepower vs. Price

```python
sns.scatterplot(x='horsepower', y='price', data=df, color='green', edgecolor='w')
plt.title('Horsepower vs. Price')
plt.xlabel('Horsepower')
plt.ylabel('Price')
plt.grid(True)

plt.show()
```

## iii- Fuel Type Distribution

```python
sns.countplot(x='fuel-type', data=df, palette='Set2')
plt.title('Fuel Type Distribution')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.grid(True, axis='y')

plt.show()
```
